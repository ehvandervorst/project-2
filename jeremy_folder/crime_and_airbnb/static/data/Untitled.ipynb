```python
import pandas as pd

from sodapy import Socrata
```

```python
client = Socrata("data.cityofnewyork.us", "w5ANkINfDguTjEbedyzkLA1HW")

results = client.get("5uac-w243", limit=3000)

crime_df = pd.DataFrame.from_records(results)
crime_df.head()
```

```python
crime_df_filtered = crime_df[["rpt_dt", "addr_pct_cd", "ofns_desc", "loc_of_occur_desc", "prem_typ_desc", "latitude", "longitude", "lat_lon"]].copy()
```

```python
crime_df_filtered
```

```python
crime_df_filtered = crime_df_filtered.dropna(subset=["lat_lon"])
crime_df_filtered
```
